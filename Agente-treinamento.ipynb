```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# Criando o ambiente: 10x10 grid
tamanho = 10
ambiente = np.zeros((tamanho, tamanho))

# Definindo a posição inicial do agente
pos_agente = [0, 0]

# Definindo o objetivo
objetivo = [9, 9]

# Marcando o objetivo no grid
ambiente[objetivo[0], objetivo[1]] = 2  # 2 representa o objetivo

def mostrar_ambiente(posicao):
    plt.imshow(ambiente, cmap='Blues')
    plt.scatter(posicao[1], posicao[0], color='red', s=200, label='Agente')
    plt.scatter(objetivo[1], objetivo[0], color='green', s=200, label='Objetivo')
    plt.legend()
    plt.grid(False)
    plt.show()

# Visualizando o ambiente inicial
mostrar_ambiente(pos_agente)
```

```python
def mover_agente(pos_agente, objetivo):
    nova_pos = pos_agente.copy()
    if pos_agente[0] < objetivo[0]:
        nova_pos[0] += 1  # move para baixo
    elif pos_agente[0] > objetivo[0]:
        nova_pos[0] -= 1  # move para cima
    elif pos_agente[1] < objetivo[1]:
        nova_pos[1] += 1  # move para direita
    elif pos_agente[1] > objetivo[1]:
        nova_pos[1] -= 1  # move para esquerda
    return nova_pos

# Simulação do movimento até o objetivo
posicao_atual = pos_agente.copy()

while posicao_atual != objetivo:
    mostrar_ambiente(posicao_atual)
    posicao_atual = mover_agente(posicao_atual, objetivo)
    time.sleep(0.5)  # Pequeno delay para visualização

mostrar_ambiente(posicao_atual)
print("Objetivo alcançado!")
```
